```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate F(t) based on the given equation
def calculate_force(m, L, g, theta, t):
    # Estimate velocity (first derivative of theta)
    theta_dot = np.gradient(theta, t)  # Velocity (rad/s)
    
    # Estimate acceleration (second derivative of theta)
    theta_ddot = np.gradient(theta_dot, t)  # Acceleration (rad/s^2)
    
    # Calculate the external force F(t)
    F_t = m * L * (theta_ddot + (g / L) * theta)
    
    return F_t

def main():
    try:
        # Input values for mass, length, and gravity
        m = float(input("Enter the mass of the pendulum (kg): "))
        L = float(input("Enter the length of the pendulum (m): "))
        g = float(input("Enter the gravitational constant (usually 9.81 m/s^2): "))
        
        # Get the time array (or generate one)
        t_start = float(input("Enter the start time (s): "))
        t_end = float(input("Enter the end time (s): "))
        num_points = int(input("Enter the number of time points: "))
        
        t = np.linspace(t_start, t_end, num_points)  # Time array
        
        # Get the position (theta) values from a mathematical model (simple harmonic motion in this case)
        theta_0 = float(input("Enter the amplitude of oscillation (theta_0 in radians): "))
        omega = np.sqrt(g / L)  # Natural angular frequency
        theta = theta_0 * np.sin(omega * t)  # Generate theta(t) using simple harmonic motion
        
        # Calculate F(t)
        F_t = calculate_force(m, L, g, theta, t)
        
        # Display the first few values of F(t)
        print("First few values of F(t):", F_t[:10])
        
        # Plot the results
        plt.figure(figsize=(10, 6))
        plt.plot(t, F_t, label="F(t) (External Force)")
        plt.xlabel('Time (s)')
        plt.ylabel('Force (N)')
        plt.title('External Forcing Function F(t) Over Time')
        plt.legend()
        plt.grid(True)
        plt.show()
    
    except ValueError:
        print("Error: Please enter valid numerical inputs.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    main()
```
